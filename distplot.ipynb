```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
data_1 = np.random.normal(0, 1, size=100)
data_2 = np.random.normal(10, 2, size=100)
values = pd.Series(np.concatenate([data_1, data_2]))
sns.distplot(values)
```

```python
sns.distplot(values, bins=100, color="pink")
```
